# Problemas de distribución de probabilidad

## Distribución binomial

En teoría de la probabilidad y estadística, la distribución binomial es la distribución de probabilidad discreta que da sólo dos resultados posibles en un experimento; por ejemplo, al lanzar una moneda, puede salir cara o cruz.

```python
from scipy.stats import binom

# Definir todas las funciones de probabilidad relacionadas con esta distribución.

def dbinom(x, size, prob = 0.5):
    """
    Calcula la estimación puntual de la distribución binomial.
    """
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)

    return result

def pbinom(q, size, prob = 0.5):
    """
    Calcula el acumulado de la distribución binomial.
    """
    result = binom.cdf(k = q, n = size, p = prob, loc = 0)

    return result

def qbinom(p, size, prob = 0.5):
    """
    Calcula la función cuantil a partir de la distribución binomial.
    """
    result = binom.ppf(q = p, n = size, p = prob, loc = 0)

    return result

def rbinom(n, size, prob = 0.5):
    """
    Genera variables aleatorias a partir de la distribución binomial.
    """
    result = binom.rvs(n = size, p = prob, size = n)

    return result
```

```python
# Distribución binomial(10, 0.2), dónde n = 10 es el número de ensayos y p = 0.2 es la probabilidad de éxito en un solo ensayo
import numpy as np

np.random.seed(42)

print(f"Probability that a Binomial(10, 0.2) takes the value 2: {dbinom(2, size = 10, prob = 0.2)}")
print(f"Probability that a Binomial(10, 0.2) will take a value less than 2: {pbinom(2, size = 10, prob = 0.2)}")
print(f"Which value of a Binomial(10, 0.2) has a cumulative probability of 0.9? {qbinom(0.9, size = 10, prob = 0.2)}")
print(f"Generate 2000 random values from a Binomial(10, 0.2) distribution: {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
import matplotlib.pyplot as plt

np.random.seed(42)

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

fig, axis = plt.subplots(3, 1, figsize = (5, 7))

axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# TODO
from scipy.stats import norm, uniform

# Distribucion Normal
def dnorm(x, loc=0, scale=1):
    """
    Calcula la estimación puntual de la distribucion normal.
    """
    result = norm.pdf(x, loc=loc, scale=scale)
    return result

def pnorm(q, loc=0, scale=1):
    """
    Calcula el acumulado de la distribucion normal.
    """
    result = norm.cdf(q, loc=loc, scale=scale)
    return result

def qnorm(p, loc=0, scale=1):
    """
    Calcula la función cuantil a partir de la distribucion normal.
    """
    result = norm.ppf(p, loc=loc, scale=scale)
    return result

def rnorm(n, loc=0, scale=1):
    """
    Genera variables aleatorias a partir de la distribucion normal.
    """
    result = norm.rvs(loc=loc, scale=scale, size=n)
    return result

# Distribución Uniforme
def dunif(x, loc=0, scale=1):
    """
    Calcula la estimación puntual de la distribucion uniforme.
    """
    result = uniform.pdf(x, loc=loc, scale=scale)
    return result

def punif(q, loc=0, scale=1):
    """
    Calcula el acumulado de la distribucion uniforme.
    """
    result = uniform.cdf(q, loc=loc, scale=scale)
    return result

def qunif(p, loc=0, scale=1):
    """
    Calcula la función cuantil a partir de la distribucion uniforme.
    """
    result = uniform.ppf(p, loc=loc, scale=scale)
    return result

def runif(n, loc=0, scale=1):
    """
    Genera variables aleatorias a partir de la distribucion uniforme.
    """
    result = uniform.rvs(loc=loc, scale=scale, size=n)
    return result
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm, uniform

np.random.seed(42)

# Distribución Normal

loc = 0
scale = 1
data = np.arange(-5, 5, 0.01)
pdf = norm.pdf(data, loc=loc, scale=scale)
cdf = norm.cdf(data, loc=loc, scale=scale)
ppf_data = np.linspace(0.01, 0.99, 100)
ppf = norm.ppf(ppf_data, loc=loc, scale=scale)

fig, axis = plt.subplots(3, 1, figsize=(5, 7))
axis[0].plot(data, pdf, "r-")
axis[0].set_title("PDF")
axis[1].plot(data, cdf, "b-")
axis[1].set_title("CDF")
axis[2].plot(ppf_data, ppf, "g-")
axis[2].set_title("PPF")

plt.tight_layout()
plt.show()

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()


```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm, uniform

np.random.seed(42)

# Distribución Uniforme
low = 0
high = 1
data_uniform = np.arange(0, 1, 0.01)
pdf_uniform = uniform.pdf(data_uniform, loc=low, scale=high-low)
cdf_uniform = uniform.cdf(data_uniform, loc=low, scale=high-low)
ppf_data_uniform = np.linspace(0.01, 0.99, 100)
ppf_uniform = uniform.ppf(ppf_data_uniform, loc=low, scale=high-low)

fig, axis = plt.subplots(3, 1, figsize=(5, 7))
axis[0].plot(data_uniform, pdf_uniform, "r-")
axis[0].set_title("PDF")
axis[1].plot(data_uniform, cdf_uniform, "b-")
axis[1].set_title("CDF")
axis[2].plot(ppf_data_uniform, ppf_uniform, "g-")
axis[2].set_title("PPF")

plt.tight_layout()
plt.show()
```
